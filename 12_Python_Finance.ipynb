##2. Índices

```python
inv.get_indices_list('brazil')
```

```python
inv.get_index_recent_data('Bovespa', country='brazil')
```

```python
ibov = inv.get_index_historical_data('Bovespa', country='brazil', from_date='01/01/1900', to_date='12/04/2021')
ibov
```

```python
ibov_dolarizado = inv.get_index_historical_data('Ibovespa USD', country='brazil', from_date='01/01/1900', to_date='12/04/2021')
ibov_dolarizado
```

```python
ifix = inv.get_index_historical_data('BM&FBOVESPA Real Estate IFIX', country='brazil', from_date='01/01/1900', to_date='12/04/2021')
ifix
```

```python
indices = pd.DataFrame()
data_inicio = ifix.index[0]

indices['ifix'] = ifix['Close'] / ifix['Close'].loc[data_inicio]
indices['ibov'] = ibov['Close'] / ibov['Close'].loc[data_inicio]
```

```python
indices.plot();
```
