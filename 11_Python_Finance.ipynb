O **BOVA11** é o principal ETF do Brasil, gerido pela BlackRock, uma empresa americana que está entre as maiores gestoras do mundo, e contém em carteira as ações mais negociadas do Brasil.

O **BOVV11** é o ETF também espelhado no índice Ibovespa, e também composto por companhias que detém uma fatia de cerca de 80% do número de negócios e volume financeiro da B3, gerido pela instituição financeira Itaú Unibanco S.A.

O **BOVB11** é o ETF Bradesco Ibovespa Fundo de Índice, gerido pela BRAM – Bradesco Asset Management S.A..

O **XBOV11** é o CAIXA ETF Ibovespa Fundo de Índice, gerido pela Caixa Econômica Federal.

## 4. Comparativo

### 4.1 Comparativo (BOVA11 x BOVV11 x BOVB11 x XBOV11)

```python
comparativo1 = pd.DataFrame()
data_inicio = bovb11.index[0]

comparativo1['bovb11'] = bovb11 / bovb11.loc[data_inicio]
comparativo1['bovv11'] = bovv11 / bovv11.loc[data_inicio]
comparativo1['xbov11'] = xbov11 / xbov11.loc[data_inicio]
comparativo1['bova11'] = bova11 / bova11.loc[data_inicio]

comparativo1
```

```python
comparativo1.plot();
```

## 4.2 Comparativo (IVVB11 x BOVA11)

```python
comparativo2 = pd.DataFrame()
data_inicio2 = ivvb11.index[0]

comparativo2['ivvb11'] = ivvb11 / ivvb11.loc[data_inicio2]
comparativo2['bova11'] = bova11 / bova11.loc[data_inicio2]

comparativo2
```

```python
comparativo2.plot();
```
